```python
import pandas as pd # for data analysis 
import matplotlib.pyplot as plt # for data visulisation
```

```python
data = pd.read_csv('../../data/cso-populationbyage.csv')
```

```python
# print the head 
data.head()
```

```python
# data columns and types
data.dtypes
```

```python
# data types
data.dtypes

data['Administrative Counties'].unique()
```

```python
# seperate out monaghan
Monaghan_data = data[data['Administrative Counties'] == 'Monaghan County Council']
Monaghan_data.head()
```

```python
Monaghan_data = Monaghan_data.copy()

total_population = Monaghan_data.loc[
    Monaghan_data['Single Year of Age'] == 'All ages', 'VALUE'
].iloc[0]

# default to NaN
Monaghan_data['Age Population Percentage'] = None  

# fill only for real age groups
Monaghan_data.loc[
    Monaghan_data['Single Year of Age'] != 'All ages',
    'Age Population Percentage'
] = (
    Monaghan_data.loc[
        Monaghan_data['Single Year of Age'] != 'All ages', 'VALUE'
    ] / total_population * 100
)
```

```python
# plot the percentages by age from monaghan 
plt.figure(figsize=(12, 6))
plt.plot(
    Monaghan_data['Single Year of Age'],
    Monaghan_data['Age Population Percentage'],
    marker='o'
)
plt.title('Population Percentage by Age in Monaghan')
plt.xlabel('Age')
plt.ylabel('Population Percentage (%)')
plt.xticks(rotation=90)
plt.grid()
plt.tight_layout()
plt.show()
```

```python
# group by age groups and sum the population percentages
age_groups = data.groupby('Single Year of Age')['VALUE'].sum().reset_index()
age_groups = age_groups[age_groups['Single Year of Age'] != 'All ages']
print(age_groups)
```

```python
for counties in data['Administrative Counties'].unique():
    if counties == 'All counties':
        continue
    county_data = data[data['Administrative Counties'] == counties].copy()
    total_population = county_data.loc[
        county_data['Single Year of Age'] == 'All ages', 'VALUE'
    ].iloc[0]
    county_data['Age Population Percentage'] = None  
    county_data.loc[
        county_data['Single Year of Age'] != 'All ages',
        'Age Population Percentage'
    ] = (
        county_data.loc[
            county_data['Single Year of Age'] != 'All ages', 'VALUE'
        ] / total_population * 100
    )
    
    plt.figure(figsize=(12, 6))
    plt.plot(
        county_data['Single Year of Age'],
        county_data['Age Population Percentage'],
        marker='o'
    )
    plt.title(f'Population Percentage by Age in {counties}')
    plt.xlabel('Age')
    plt.ylabel('Population Percentage (%)')
    plt.xticks(rotation=90)
    plt.grid()
    plt.tight_layout()
    plt.show()
```
