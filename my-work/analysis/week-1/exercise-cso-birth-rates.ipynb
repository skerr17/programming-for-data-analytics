```python
import pandas as pd # for data analysis 
import matplotlib.pyplot as plt # for data visulisation  
```

```python
# Read the CSV file into a DataFrame
data = pd.read_csv('../../data/projectedbirths-cso.csv')
```

```python
# Display the first few rows of the DataFrame
data.head()
```

```python
# dtype of each column
data.dtypes
```

```python
# plot the data in a bar chart x = 'Year', y = 'Projected Births'

plt.bar(data['Year'], data['VALUE'])
plt.xlabel('Year')
plt.ylabel('Projected Births')
plt.title('Projected Births in Ireland Over Time')
plt.grid(True)
plt.show()
```

```python
plt.plot(data['Year'], data['VALUE'], marker='o')
plt.xlabel('Year')
plt.ylabel('Projected Births')
plt.title('Projected Births in Ireland Over Time')
plt.grid(True)
plt.show()
```

```python
data['MovingAvg'] = data['VALUE'].rolling(window=3).mean()
plt.plot(data['Year'], data['VALUE'], label='Actual')
plt.plot(data['Year'], data['MovingAvg'], label='3-Year Moving Avg', linestyle='--')
plt.xlabel('Year')
plt.ylabel('Projected Births')
plt.title('Projected Births with Moving Average')
plt.legend()
plt.show()
```
